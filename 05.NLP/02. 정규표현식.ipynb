# 1. 정규표현식 실습 

```python
# . 기호 
print(re.search('a.c', 'kkk')) # a하고 c인 임의의 한 글자가 없기 때문에 아무것도 안나온다. 
re.search('a.c', 'abc')
```

```python
# ? 기호 

r = re.compile('ab?c')
r.search('abbc') # 얘는 2개가 있기 때문에 안됨 
```

```python
r.search('abc')
```

```python
r.search('ac')
```

```python
# * 기호 : 와일드카드 느낌 
r = re.compile('ab*c')
r.search('a')
```

```python
r.search('ac')
```

```python
r.search('abbbbc')
```

```python
# + 기호 : 앞의 문자가 최소 1개 이상 

r = re.compile("ab+c")
r.search('ac')
```

```python
r.search('abbbbbbc')
```

```python
# ^ : 시작되는 글자를 지정 

re.search ('^a', 'bbc')
```

```python
re.search('^a', 'ab')
```

```python
# {숫자} : 해당 문자를 숫자만큼 반복 

r = re.compile('ab{2}c')
r.search('ac')
```

```python
r.search('abbc')
```

```python
# {숫자1, 숫자2} : 해당 문자를 숫자1 이상 숫자 2이하만큼 반복 

r = re.compile('ab{2,8}c')
r.search('abc')
```

```python
r.search('abbbbc')
```

```python
# [] : 문자들 중 한 개의 문자와 매치하라는 의미 
r = re.compile("[abc]") # abc는 a-c와 같다. 
r.search('zzz')
```

```python
r.search('baac')
```

```python
# 소문자 
re.search('[a-z]', 'ABC')
```

```python
re.search('[a-z]', 'AbC')

# 대문자, 숫자 [A-Z0-9]
```

```python
# [^문자] : ^ 뒤에 문자들을 모두 제외한 모든 문자를 매치 

r = re.compile('[^abc]')
r.search('a')
```

```python
r.search('d1')
```

```python
# 문자 또는 숫자 
re.search('[0-9A-Za-z]', '!@#$')
re.search('\w', '!@#$') # 위의 코드와 같은 의미임  
```

```python
# 한글과 공백을 제외한 모든 문자를 제거 
re.sub('[^ㄱ-ㅎㅏㅏ-ㅣ가-힣 ]', '', '한글과 space, 를 제외한 모든 문제를 제거 8.')
```
