```python
from konlpy.tag import Okt
```

```python
okt = Okt()
```

```python
import pandas as pd 

df = pd.read_csv('/content/zflip.csv')
df.head()
```

```python
words = list(df['title'])
temp = list(df['content'])
words = words +  temp 
words
```

```python
tokens = []
for word in words : 
  token = okt.nouns(word)
  tokens = tokens + token 
tokens
```

```python
import re 
new_tokens = []

for token in tokens : 
  item = re.sub('[A-Za-z0-9]', '', token)
  if item:
        new_tokens.append(item)
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
plt.rc('font', family='NanumBarunGothic')
```

```python

import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
zflip = nltk.Text(new_tokens, name='z플립')
plt.figure(figsize=(15,6))
zflip.plot(50)
plt.show()
```

```python
stoptext = """
  플립 갤럭시 폴드 삼성 및 제 플 사용 이번 지금 분 의 및 종 글 때 질문 가장 품 가지 생각 기 폰 플 것 종 글 수 더 점 이유 정리 모두 관심 옆 확 를 부터 한번 개 팬 중 택 
  선 확 편 도 그 안 요 위 트 팩 새 폴 과 나본 먼저 면 
  """
stop_words = stoptext.split()
new_tokens = [word for word in new_tokens if word not in stop_words]
new_tokens
```

```python

from wordcloud import WordCloud
from PIL import Image
```

```python
data = zflip.vocab().most_common(300)
path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
wc = WordCloud(
        font_path=path, relative_scaling=0.2,
        background_color='white'
).generate_from_frequencies(dict(data))
```

```python
plt.figure(figsize=(12,6))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python

mask = np.array(Image.open('/content/unnamed.png'))

from wordcloud import ImageColorGenerator
image_colors = ImageColorGenerator(mask)
```

```python
wc = WordCloud(
        font_path=path, relative_scaling=0.2,
        background_color='white', mask=mask,
        min_font_size=1, max_font_size=120
).generate_from_frequencies(dict(data))
```

```python
plt.figure(figsize=(10,10))
plt.imshow(wc.recolor(color_func=image_colors), 
           interpolation='bilinear')
plt.axis('off')
plt.show()
```
